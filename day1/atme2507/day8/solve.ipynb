```python
df['join_year']=df['join_date'].dt.year
df['join_month']=df['join_date'].dt.month
```

```python
step 2
# FILTERING AND CONDITIONAL SELECTION
```

```python
it_above_75k=df[(df['department']=='IT')&(df['salary']>75000)]
it_above_75k
```

```python
before_2020=df[df['join_date']<'2020-01-01']
before_2020
```

```python
before_2017=df[df['join_date']<'2017-01-01']
before_2017
```

```python
sorted_by_ps=df.sort_values(by='performance_score',ascending=False)
sorted_by_ps.head()
```

```python
marketing_boston=df[(df['department']=='Marketing')&(df['city']=='Boston')]
marketing_boston
# list all the employees from boston in the marketing dept 
```

```python
miss_sal=df[df['salary'].isna()]
miss_sal
#list all the emploees with missing salary
```

```python
step 3
#AGGREGATION AND GROUPING
```

```python
dept_avg_sal=df.groupby('department')['salary'].mean()
dept_avg_sal
# avaearge salary by department
```

```python
city_emps=df.groupby('city')['name'].count()
city_emps
# number of employees in each city
```

```python
ps_max_per_dept=df.groupby('department')['performance_score'].max()
ps_max_per_dept
# maximum performance score by department
```

```python
# Total salary paid to employees in the Finance department
total_finance_salary = df[df['department'] == 'Finance']['salary'].sum()
total_finance_salary
```

```python
# Number of employees who joined each year
year_joinees = df.groupby('join_year')['name'].count()
year_joinees
```
