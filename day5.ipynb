```python
#implement student record operation CRUD using nested dictionaries
student_records = {}

# Function to add a new student
def add_student():
    student_id = input("Enter Student ID: ")
    # Check if student ID already exists
    if student_id in student_records:
        print("Student ID already exists.")
        return
    # Collect student details
    name = input("Enter Name: ")
    age = input("Enter Age: ")
    grade = input("Enter Grade: ")
    # Add student to the dictionary
    student_records[student_id] = {
        'name': name,
        'age': age,
        'grade': grade
    }
    print("Student added successfully.\n")

# Function to view all student records
def view_all_students():
    # Check if there are any records
    if not student_records:
        print("No student records found.\n")
        return
    print("\n--- All Student Records ---")
    # Loop through each student and print their details
    for sid, info in student_records.items():
        print(f"ID: {sid}, Name: {info['name']}, Age: {info['age']}, Grade: {info['grade']}")
    print()

# Function to search for a student by ID
def search_student():
    student_id = input("Enter Student ID to search: ")
    # Retrieve student record using get()
    student = student_records.get(student_id)
    if student:
        print(f"ID: {student_id}, Name: {student['name']}, Age: {student['age']}, Grade: {student['grade']}\n")
    else:
        print("Student not found.\n")

# Function to update an existing student record
def update_student():
    student_id = input("Enter Student ID to update: ")
    # Check if student exists
    if student_id not in student_records:
        print("Student not found.\n")
        return
    # Ask for new values; leave blank to keep current
    name = input("Enter new Name (leave blank to keep current): ")
    age = input("Enter new Age (leave blank to keep current): ")
    grade = input("Enter new Grade (leave blank to keep current): ")

    # Update only the fields that are provided
    if name:
        student_records[student_id]['name'] = name
    if age:
        student_records[student_id]['age'] = age
    if grade:
        student_records[student_id]['grade'] = grade
    print("Student record updated.\n")

# Function to delete a student record
def delete_student():
    student_id = input("Enter Student ID to delete: ")
    # Check if student exists
    if student_id in student_records:
        del student_records[student_id]  
        print("Student record deleted.\n")
    else:
        print("Student not found.\n")

def main():
    while True:
        # Display menu options
        print("------ Student Management System ------")
        print("1. Add Student Record")
        print("2. View All Students")
        print("3. Search Student by ID")
        print("4. Update Student Record")
        print("5. Delete Student Record")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")

        # Call appropriate function based on user input
        if choice == '1':
            add_student()
        elif choice == '2':
            view_all_students()
        elif choice == '3':
            search_student()
        elif choice == '4':
            update_student()
        elif choice == '5':
            delete_student()
        elif choice == '6':
            print("Exiting Student Management System. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.\n")

# Run the program
if _name_ == "_main_":
    main()
```
